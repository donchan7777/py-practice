## matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = [12.4, 10.9, 12.2, 12.0, 11.4, 11.7, 11.5, 11.4, 11.4, 12.5]
y = [200, 400, 600, 800, 1000, 1200, 1400, 1600, 1800, 2000]

plt.plot(y, x,
        color='red',
        lw=1,
        )
plt.xlabel('m')
plt.ylabel('time')
plt.title('NGT')
plt.xticks([200,400,600,800,1000,1200,1400,1600,1800,2000])
```

```python
x = np.linspace(-10, 10, 100)
z = 1 / (1 + np.exp(-x))
plt.plot(x, z)
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
df = df[~df['release_date'].isna()]
```

```python
df['year'] = df['release_date'].apply(lambda x: int(x[:4]))
```

```python
plt.plot(df.groupby('year').count()['id'])
plt.xlabel('year')
plt.ylabel('count')
```

```python
df = df[df['revenue']!=0]
plt.plot(df.groupby('year').median()['revenue'], label='median')
plt.plot(df.groupby('year').mean()['revenue'], label='mean')
plt.plot(df.groupby('year')['revenue'].max(), label='max')
plt.legend()
```

```python
df = pd.read_csv("tmdb_5000_movies.csv")
```

```python
plt.scatter(df['revenue'], df['budget'], alpha=0.2)
```

```python
df = df[(df['revenue'] != 0) & (df['budget'] != 0)]
lang_group = df.groupby('original_language').mean()[['budget', 'revenue', 'vote_count']]
c = np.arange(len(lang_group))
plt.scatter('revenue', 'budget', s='vote_count', c=c, data=lang_group)
             
for idx, row in lang_group.iterrows():
    plt.annotate(idx, (row['revenue'], row['budget']))

fig.tight_layout()
```
