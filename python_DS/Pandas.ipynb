## Filter

```python
data = {
    'name': ['John', 'Zack', 'Emily'],
    'sex': ['male', 'male', 'female'],
    'age': [22, 30, 32]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df[[True, False, True]]
```

```python
filter_s = df['age'] >= 30
```

```python
df[filter_s]
```
