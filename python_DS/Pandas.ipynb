## Pivotテーブルの作り方

```python
data = {'Date': ['Jan-1', 'Jan-2', 'Jan1', 'Jan-2', 'Jan-2', 'Jan-2'],
       'User': ['Emily', 'John', 'Nick', 'Kevin', 'Emily', 'John'],
       'Method': ['Cart', 'Cart', 'Cash', 'Cart', 'Cash', 'Cash'],
       'Price': [100, 250, 200, 460, 200, 130]}
df = pd.DataFrame(data)
df
```

## DataFrameの保存

```python
df = pd.read_csv('tmdb_5000_movies.csv')

df = df[~(df['budget'] == 0) | (df['revenue'] == 0)]

df['profit'] = df.apply(lambda row: row['revenue'] - row['budget'], axis=1)

df = df.sort_values('profit', ascending=False)
```

```python
df.head(2)
```

```python
df.to_csv('tmdb_5000_profit.csv')
```
